#### Exercise 1

Given the code below    

    text = 'Business Analytics'

What will the following expressions produce?
    
    text[4]
    text[2:4]
    text[:4]
    text[4:]
    text[-2]
    text[-4:-2]
    text[-2:]
    text[:-2]
    

#### Exercise 1

```python
#Given the code below

text = 'Business Analytics'

#What will the following expressions produce?

letter = text[4]
letter1 = text[2:4]
letter2 = text[2:4]
letter3 = text[4:]
letter4 = text[-2]
letter5 = text[-4:-2]
letter6 = text[-2:]
letter7 = text[:-2]

print("The following expression will produce:")

print("text[4]: " + letter)
print("text[2:4]: " +letter1)
print("text[2:4]: " + letter2)
print("text[4:]: " + letter3)
print("text[-2]: " + letter4)
print("text[-4:-2]: " + letter5)
print("text[-2:]: " + letter6)
print("text[:-2]: " + letter7)
```

#### Exercise 2

    phrase = 'One Ring to rule them all'

What string slicing expression gives out the following words:

    'One '


## CHALLENGE


Write a program which repeatedly reads a **full name**, then does the following:
- strips the name of any trailing spaces
- capitalizes the first letter
- puts the rest into smaller letters


```python

while True:
    name = input("Enter Full Name(Enter x to exit): ")
    if name.lower() == "x":
      break
    print(name.strip().capitalize())
```
