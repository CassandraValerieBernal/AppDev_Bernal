##What are the constants in the math module?
The CONSTANTS in math module are:

math.pi -- to access the value of pi

math.e -- to access the value Euler's Number

##What's does math.exp do? What about math.pow?

math.exp computes the exponential of the number ex:
math.exp(x) equivalent to e^x ---

### exp(x, /)
        Return e raised to the power of x.

math.pow computes the power of a number. It takes two arguements, the base and the exponent.

### pow(x, y, /)
        Return x**y (x to the power of y).

```python
%%writefile BERNAL.py

book_library = {}

# Add books directly to the book_library dictionary
book_library["Alice in Wonderland"] = "Lewis Carroll"
book_library["Animal Farm"] = "George Orwell"
book_library["To Kill a Mockingbird"] = "Harper Lee"
book_library["Pride and Prejudice"] = "Jane Austen"

def display_books():
    """
    Display the books in the library.
    """
    if not book_library:
        print("No Books in the Library.")
    else:
        print("Books in the Library:")
        for title, author in book_library.items():
            print(f"- Title: {title}, Author: {author}")
```
