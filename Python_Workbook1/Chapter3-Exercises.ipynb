#### Exercise 2

Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    
[PEW] E3.2

````python
hours = int(input('Enter Hours: '))
rate = int(input('Enter Rate: '))

if hours > 40:
  ot = hours - 40
  nrate = ot*(rate*1.5)
  npay = 40*rate + nrate
  print('Pay: ', npay)

else:
  pay = hours * rate
  print('Pay: ', pay)
````

## CHALLENGE

Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

    Score           Grade
    [ 97, 100]      1.0
    [ 93,  97)		1.25
    [ 89,  93)		1.5
    [ 85,  89)		1.75
    [ 81,  85)		2.0
    [ 77,  81)		2.25
    [ 74,  77)		2.5
    [ 0,   74)		3.0

Adapted from [PEW] E3.3
<br>

````python
grade = int(input('Enter Score: '))

if grade >= 0 and grade <=74:
  print ('Your Grade is: 3.0')
elif grade >= 74 and grade <=77:
  print ('Your Grade is: 2.5')
elif grade >= 77 and grade <=81:
  print ('Your Grade is: 2.25')
elif grade >= 81 and grade <=85:
  print ('Your Grade is: 2.0')
elif grade >= 85 and grade <=89:
  print ('Your Grade is: 1.75')
elif grade >= 89 and grade <=93:
  print ('Your Grade is: 1.5')
elif grade >= 93 and grade <=97:
  print ('Your Grade is: 1.25')
elif grade >= 97 and grade <=100:
  print ('Your Grade is: 1.0')
else:
  print('Invalid Score')
````
