#### Exercise 1

Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    
[PEW] E4.6

```python
hours = int(input('Enter Hours: '))
rate = int(input('Enter Rate: '))

def computepay (hours, rate):
  if hours > 40:
    ot = hours - 40
    nrate = ot*(rate*1.5)
    npay = 40*rate + nrate
    print('Pay: ', npay)
  else:
    pay = hours * rate
    print('Pay: ', pay)


computepay(hours, rate)




```

## CHALLENGE

#### Challenge

Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

    Score           Grade
    [ 97, 100]      1.0
    [ 93,  97)		1.25
    [ 89,  93)		1.5
    [ 85,  89)		1.75
    [ 81,  85)		2.0
    [ 77,  81)		2.25
    [ 74,  77)		2.5
    [ 0,   74)		3.0

    Run the program repeatedly to test the various different values for input.
    
Adapted from [PEW] E4.6
<br>

```python
while True:
    input = input('Enter Score (type "done" to finish): ')

    if input.lower() == "done":
        break

    try:
        grade = int(input)
        if grade < 0 or grade > 100:
            print('Invalid Score')
            continue
    except ValueError:
        print('Invalid input. Please enter a valid integer score.')
        continue

    def computegrade(grade):
        if grade >= 0 and grade <= 74:
            print('Your Grade is: 3.0')
        elif grade >= 75 and grade <= 77:
            print('Your Grade is: 2.5')
        elif grade >= 78 and grade <= 81:
            print('Your Grade is: 2.25')
        elif grade >= 82 and grade <= 85:
            print('Your Grade is: 2.0')
        elif grade >= 86 and grade <= 89:
            print('Your Grade is: 1.75')
        elif grade >= 90 and grade <= 93:
            print('Your Grade is: 1.5')
        elif grade >= 94 and grade <= 97:
            print('Your Grade is: 1.25')
        elif grade >= 98 and grade <= 100:
            print('Your Grade is: 1.0')

    computegrade(grade)
print("Done!")

```
